```python
import os
import cv2
import numpy as np
from sklearn.cluster import MiniBatchKMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, f1_score,classification_report, confusion_matrix
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random

```

```python
print("Let us first read the train and test files")

paths = {}

paths['trainImages'] = "./train/"
paths['testImages'] = "./test/"

def list_dir(path_dir):
    f = []
    for (dirpath, dirnames, filenames) in os.walk(path_dir):
        f.extend(dirnames)
        break
    return f

def get_label_path(path_dir):
    path_label = list_dir(path_dir)
    listImages ={}
     
    for label in path_label:
        
        # Import Data from directories
        for filename in os.listdir(path_dir + label):
            if filename.endswith(".jpg"):
                listImages[path_dir + label + '/'+ filename] = label
    return listImages

train = get_label_path(paths['trainImages'])
test = get_label_path(paths['testImages'])
```

```python
print("We create a SIFT object detector and descriptor")
print("We compute the SIFT descriptors for all the train images and subsequently build a numpy array with all the descriptors stacked together")

def numFeautresIteration(valueNumFeatures):
    SIFTdetector = cv2.xfeatures2d.SIFT_create(nfeatures=valueNumFeatures)
    
    Train_descriptors = []
    Train_label_per_descriptor = []
    
    for filename,labels in zip(train.keys(), train.values()):
        ima=cv2.imread(filename)
        gray=cv2.cvtColor(ima,cv2.COLOR_BGR2GRAY)
        kpt,des=SIFTdetector.detectAndCompute(gray,None)
        Train_descriptors.append(des)
        Train_label_per_descriptor.append(labels)
    
    D=np.vstack(Train_descriptors)
    return D, Train_descriptors, SIFTdetector
```

```python
print("We now compute a k-means clustering on the descriptor space")
print("And, for each train image, we project each keypoint descriptor to its closest visual word. We represent each of the images with the frequency of each visual word.")
def numClustersIteration(k, D, Train_descriptors):
    codebook = MiniBatchKMeans(n_clusters=k, verbose=False, batch_size=k * 20,compute_labels=False,reassignment_ratio=10**-4,random_state=42)
    codebook.fit(D)
    
    visual_words=np.zeros((len(Train_descriptors),k),dtype=np.float32)
    for i in range(len(Train_descriptors)):
        words=codebook.predict(Train_descriptors[i])
        visual_words[i,:]=np.bincount(words,minlength=k)
        
    return visual_words, codebook
```

```python
print("We build a k-nn classifier and train it with the train descriptors")
print("We end up computing the test descriptors and compute the accuracy of the model")
def numNeighIteration(metric, neigh, visual_words, codebook, SIFTdetector):
    knn = KNeighborsClassifier(n_neighbors=neigh,n_jobs=-1,metric=metric)
    knn.fit(visual_words, list(train.values())) 
    
    
    visual_words_test=np.zeros((len(test.keys()),k),dtype=np.float32)
    testKeys = list(test.keys())
    for i in range(len(testKeys)):
        filename=testKeys[i]
        ima=cv2.imread(filename)
        gray=cv2.cvtColor(ima,cv2.COLOR_BGR2GRAY)
        kpt,des=SIFTdetector.detectAndCompute(gray,None)
        words=codebook.predict(des)
        visual_words_test[i,:]=np.bincount(words,minlength=k)
    
    
    accuracy = 100*knn.score(visual_words_test, list(test.values()))
    indexMetric = ['euclidean','manhattan','chebyshev','minkowski'].index(metric)
    indexNeigh = int((neigh/4)-6)
    mat_accuracy[indexMetric][indexNeigh] = accuracy
    
    return(mat_accuracy)
```

```python
matK = range(4,8)
matNeigh = range(24,35,3)
matNumFeatures = range(1,5)
matMetrics = ['euclidean','manhattan','chebyshev','minkowski']
matMetricIndex = range(1,5)
for valueNumFeatures in matNumFeatures:
    valueNumFeatures = valueNumFeatures*60
    D ,Train_descriptors, SIFTdetector = numFeautresIteration(valueNumFeatures)
    
    mat_accuracy = np.zeros((len(matMetrics), len(list(matNeigh))))
    for k in matK:
        k = 2**k
        visual_words, codebook =  numClustersIteration(k, D, Train_descriptors)
       
        for metric in matMetrics:
            for neigh in matNeigh:
                
                mat_accuracy = numNeighIteration( metric, neigh, visual_words, codebook, SIFTdetector)
        
        X, Y = np.meshgrid( list(matNeigh), list(range(1,5)))

        fig = plt.figure()
        ax = fig.gca(projection='3d')
        ax.set_title(("Num K = ", k, " Num Features = ", valueNumFeatures, "Max Accuracy = ", np.amax(mat_accuracy)))
        ax.set_yticklabels(['euclidean','','manhattan','','chebyshev','','minkowski'])
        ax.set_xlabel("Neighbours")
        ax.set_zlabel("Accuracy")
        ax.view_init(elev=45, azim=None)
        ax.plot_surface(X, Y, mat_accuracy,cmap='plasma')
        print("Num K = ", k, " Num Features = ", valueNumFeatures, "Max Accuracy = ", np.amax(mat_accuracy))
                
```

<H3>Unary Parameter Dependent</H3>

In order to compute the high time complexity emprovements, we implemented a method that only expect to return a unical result given the best combination of parameters, that will help us to manage the results of the Dense SIFT.

```python
def fitUnaryParameterDependent(metric, neigh, visual_words, codebook, SIFTdetector, step):
    knn = KNeighborsClassifier(n_neighbors=neigh,n_jobs=-1,metric=metric)
    knn.fit(visual_words, list(train.values())) 
    
    
    visual_words_test=np.zeros((len(test.keys()),k),dtype=np.float32)
    testKeys = list(test.keys())
    for i in range(len(testKeys)):
        filename=testKeys[i]
        ima=cv2.imread(filename)
        gray=cv2.cvtColor(ima,cv2.COLOR_BGR2GRAY)
        kps = []
        for x in range(0, gray.shape[1], step):
            for y in range(0, gray.shape[0], step):
                scale = step * random.uniform(1, 3)
                kp = cv2.KeyPoint(x, y, scale)
                kps.append(kp)
        kpt,des=SIFTdetector.compute(gray,kps)
        words=codebook.predict(des)
        visual_words_test[i,:]=np.bincount(words,minlength=k)
    
    accuracy = 100*knn.score(visual_words_test, list(test.values()))
    return(accuracy)

```

<H1> Dense Sift Emprovement</H1>
The difference between SIFT and Dense SIFT is that with dense SIFT you get a  descriptor at every location, while with normal SIFT you get a descriptions at the locations determined by Lowe's algorithm.


```python
def denseSiftDescriptors(valueNumFeatures, step):
    SIFTdetector = cv2.xfeatures2d.SIFT_create(nfeatures=valueNumFeatures)
    Train_descriptors = []
    Train_label_per_descriptor = []
    
    for filename,labels in zip(train.keys(), train.values()):
        ima=cv2.imread(filename)
        gray=cv2.cvtColor(ima,cv2.COLOR_BGR2GRAY)
        kps = []
        for x in range(0, gray.shape[1], step):
            for y in range(0, gray.shape[0], step):
                scale = step * random.uniform(1, 3)
                kp = cv2.KeyPoint(x, y, scale)
                kps.append(kp)
        kpt,des=SIFTdetector.compute(gray,kps)
        Train_descriptors.append(des)
        Train_label_per_descriptor.append(labels)
    
    D=np.vstack(Train_descriptors)
    return D, Train_descriptors, SIFTdetector

valueNumFeatures = 0
step = 16
k = 96
metric = "euclidean"
neigh = 30
D ,Train_descriptors, SIFTdetector = denseSiftDescriptors(valueNumFeatures, step)
visual_words, codebook =  numClustersIteration(k, D, Train_descriptors)
mat_accuracy = fitUnaryParameterDependent( metric, neigh, visual_words, codebook, SIFTdetector, step)
print("result" ,mat_accuracy)
```
